```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df = df[['Annual Income (k$)','Spending Score (1-100)']]
df.head()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model.fit(df)
```

```python
import seaborn as sns
sns.scatterplot(df,x='Annual Income (k$)',y = 'Spending Score (1-100)',hue= model.predict(df),color='rgbu')
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=5)
model.fit(df)
sns.scatterplot(df,x='Annual Income (k$)',y = 'Spending Score (1-100)',hue= model.predict(df),color='rgbu')
```
